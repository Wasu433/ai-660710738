ตั้งค่ามาตาฐานของเพศให้ใช้คำในคอลัมน์เหมือนๆกัน

```python
def standardize_gender(sex):
    if sex in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    elif sex in ['หญิง', 'ญ', 'Female', 'F']:
        return 'หญิง'
    else:
        return 'ไม่ระบุ'
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_gender)
```

```python
df
```

ตั้งค่ามาตาฐานของคณะให้ใช้คำในคอลัมน์เหมือนๆกัน


```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวกรรม', 'วิศวฯ', 'วิศวกรรมศาสตร์']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

กำหมดให้ค่า GPA น้อยกว่า 0 จะset ให้เป็น0 ถ้าเกิน4จะset เป็น 4 ส่วนที่เป็น NaN จะใช้ค่าเฉลี่ยแทนที่


```python
df.loc[df['GPA'] < 0, 'GPA'] = 0.00
df.loc[df['GPA'] > 4, 'GPA'] = 4.00
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
df
```

ใช้ค่าเฉลี่ยของส่วนสูงทั้งหมดมาแทนที่คนที่เป็น NaN

```python
mean_height = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(mean_height)
```

```python
df
```

ใช้ค่าเฉลี่ยของน้ำหนักทั้งหมดมาแทนที่คนที่เป็น NaN

```python
mean_weight = df['น้ำหนัก_kg'].mean()
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].fillna(mean_weight)
```

```python
df
```

ใช้ค่าเฉลี่ยของชั่วโมงเรียนทั้งหมดมาแทนที่คนที่เป็น NaN

```python
mean_time = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_time)
```
